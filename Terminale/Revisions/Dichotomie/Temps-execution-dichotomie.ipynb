## L'algorithme de recherche séquentielle

L'implémentation Python de l'algorithme de recherche séquentielle dans une liste est donné ci-dessous :

````python
def sequentielle(liste, x):
    """fontion de recherche séquentielle de x dans liste
    Renvoie True si x est dans la liste et False sinon

    Entrée :
        liste : une liste de nombres
        x : un entier

    Sortie : 
        indice : un entier

    Exemples d'exécution :
    >>> sequentielle([1, 3, 8],3)
    1
    >>> sequentielle([1, 3, 8],2)
    -1
    """
    indice = False
    for i in range(0, len(liste), 1):
        if liste[i] == x:
            indice = True
    return indice
````

## L'algorithme de recherche par dichotomie itérative

L'implémentation Python de l'agorithme de recherche par dichotomie itérative dans une liste est donné ci-dessous :

````python
def dichotomie(liste,x):
    """fonction de recherche de x dans liste par dichotomie
    Renvoie True si x est dans la liste et False sinon
    Entrées :
        x, un entier
        liste, une liste d'entiers

    Sorties :
        un booléen
    
    Exemples d'exécution :
    >>> dichotomie([1, 5, 8], 8)
    True
    >>> dichotomie([1, 5, 8], 2)
    False
    """
    debut = 0
    fin = len(liste) - 1
    milieu = int((debut + fin)/2)
    
    while debut <= fin:
        if x == liste[milieu]:
            return True
        elif liste[milieu] > x:
            fin = milieu - 1
        else :
            debut = milieu + 1
        milieu = int((debut + fin)/2)

    return False
````

## L'algorithme de recherche par dichotomie recursive

L'implémentation Python de l'agorithme de recherche par dichotomie récursive dans une liste est donné ci-dessous :


````python
def dichotomie_recursive(liste, x):
    """fonction de recherche de x dans liste par dichotomie recursive
    Renvoie True si x est dans la liste et False sinon
    Entrées :
        x, un entier
        liste, une liste d'entiers

    Sorties :
        un booléen
    
    Exemples d'exécution :
    >>> dichotomie_recursive([1, 5, 8], 8)
    True
    >>> dichotomie_recursive([1, 5, 8], 2)
    False
    """
    
    if len(liste)==0 :
        return False
    m = len(liste)//2
    if liste[m] == x :
        return True
    elif liste[m] > x :
        return dichotomie_recursive(liste[:m], x)
    else :
        return dichotomie_recursive(liste[m+1:], x)
````

## Comparaison des algorithmes de recherche séquentielle, de recherche dichotomique en version itérative et de recherche dichotomique en version récursive

La fonction $trace\_3$ permet de superposer trois courbes sur le même graphique.

> Positionnez vous sur la cellule suivante, puis cliquez sur `Run`.

````python
def trace_3(x, y1, l1, y2, l2, y3, l3):
    """fonction qui trace un graphique avec X en abscisses et Y en ordonnées"""
    
    plt.scatter(x, y1, c = 'red', label=l1)
    plt.scatter(x, y2, c='blue', label=l2)
    plt.scatter(x, y3, c='green', label=l3)
    plt.legend()
    plt.show()
````

En utilisant le même principe que précédemment, on a créé une fonction $comparaison\_temps$ d'argument `n`, permettant de tracer le temps d'exécution de la fonction $dichotomie$, de la fonction $dichotomie\_recursive$ et de la fonction $sequentielle$ pour des listes de longueur $n$.

````python
def comparaison_temps(n):
    liste_temps_dichotomie = n*[0]
    liste_temps_sequentielle = n*[0]
    liste_temps_recursive = n*[0]
    liste_n = [i for i in range (1,n+1,1)]
    for i in liste_n:
        liste = remplissage_aleatoire(i)
        liste.sort()
        val = max(liste)
        debut = 1000*time.perf_counter()
        dichotomie(liste,val)
        fin = 1000*time.perf_counter()
        liste_temps_dichotomie[i-1] = fin-debut
        debut = 1000*time.perf_counter()
        sequentielle(liste,val)
        fin = 1000*time.perf_counter()
        liste_temps_sequentielle[i-1] = fin-debut
        debut = 1000*time.perf_counter()
        dichotomie_recursive(liste, val)
        fin = 1000*time.perf_counter()
        liste_temps_recursive[i-1] = fin-debut
    trace_3(liste_n,liste_temps_dichotomie, "dichotomie itérative",liste_temps_sequentielle, "sequentielle",liste_temps_recursive, "dichotomie recursive")
````
